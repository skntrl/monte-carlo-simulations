```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 200
steps = np.random.choice([-1,1], size=n_steps)
positions = np.cumsum(steps)

plt.figure(figsize=(10,3))
plt.plot(positions)
plt.xlabel("Steps")
plt.ylabel("Position")
plt.title("1D Random Walk")
plt.grid(True)
plt.show()
```

```python
# Let's plot multiple random walks together
# Change these 2 values to see how the graph looks - it is beautiful
n_walks = 5
n_steps = 1000000

plt.figure(figsize=(30,5))

for _ in range(n_walks):
    steps = np.random.choice([-1,1], size=n_steps)
    positions = np.cumsum(steps)
    plt.plot(positions, alpha=0.7)

plt.title(f"{n_walks} random walks (1D)")
plt.xlabel("Steps")
plt.ylabel("Position")
plt.grid(True)
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# 2D Walk. Try this code. Very very beautiful graphs if you change the n_steps, n_walks variables
import random
import numpy as np
import matplotlib.pyplot as plt

n_steps = 10000
n_walks = 5

plt.figure(figsize=(10,10))

for _ in range(n_walks):
    x, y = [0], [0]
    for _ in range(n_steps):
        dx, dy = random.choice([(1,0), (0,1), (-1,0), (0,-1)])
        x.append(x[-1] + dx)
        y.append(y[-1] + dy)
    plt.plot(x, y, alpha=0.5)
    plt.plot(x[-1], y[-1], 'o') # This is the final position

plt.title("2D Random Walk")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
# plt.axis("equal")
plt.show()
```

# Some 1D Browninan Motion 

Brownian motion $B(t)$ has:  

1. $B(0) = 0$
2. Steps are all independent increments & will happen in a very small interval of time
3. $B(t + s) - B(s) \approx N(0,t)$ i.e the change in position over time $t$ is Gaussian distribution with mean = 0, variance = t
4. Continuous paths - will not contain any jump

```python
# 1D brownian motion with no drift or volatality

import numpy as np
import matplotlib.pyplot as plt

T = 1.0 # Total Time
N = 10000 # Number of time steps
dt = T / N # Time step size
t = np.linspace(0, T, N)

# Browninan motion increments
dB = np.random.normal(0, np.sqrt(dt), size = N)
B = np.cumsum(dB) # cummulative sum to get position

plt.figure(figsize=(15, 4))
plt.plot(t, B)
plt.title("1D Brownian Motion")
plt.xlabel("Time")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```

```python
# Let's add Drift and Volatality

import numpy as np
import matplotlib.pyplot as plt

T = 1.0 # Total Time
N = 10000 # Number of time steps
dt = T / N # Time step size
t = np.linspace(0, T, N)

# Change these 2 values and see the graph changes
mu = 1.5 # drift
sigma = 1.5 # volatality

# Browninan motion increments
dB = np.random.normal(0, np.sqrt(dt), size = N)
B = np.cumsum(dB) # cummulative sum to get position
X = mu * t + sigma * B # this will add drift and volatality 

plt.figure(figsize=(15, 4))
plt.plot(t, X)
plt.title("1D Brownian Motion")
plt.xlabel("Time")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```
